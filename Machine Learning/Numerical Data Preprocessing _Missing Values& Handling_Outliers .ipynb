# Handling null values 

#### we have the two techniques in this
* drop 
* Imputation

#### drop 
* Dropping in handling missing values means removing rows or columns that contain missing data from the dataset.
* df.dropna()
* dropna(axis=1) → na means null values inside the column if don't mention any thing default it is rows

#### Imputation
* Imputation means imputing your null values with some other values
* Other values are → Mean → Median → Mode

* Mean Imputation → Noraml distribution → Symmetric about the mean
* Median Imputation → median will never get effect by the outliers → if outliers
* Mode Imputation → If it is discrete, numeric or categorical data → majority for categorical data

##  we can take help of two methods
* fillna
* loc → filter out the null values we can choose mean or median or mode according to your columns 

### fillna()
* directlly to fill whenever there is a  null value on mean, median or mode
* mean → df['column_name'].mean()
* df[column_name'].fillna(value='x')

# IQR  

```python
Q1 = df['tds'].quantile(0.25)
Q3 = df['tds'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
print(df['tds'].median())
lower = Q1-1.5 * IQR
upper= Q3+1.5 * IQR
print(lower)
print(upper)
```

```python
df.loc[(df['tds']<lower) | (df['tds']>upper)]
```

```python
print(sum((df['tds']<lower) | (df['tds']>upper)))
```

# Handling Outliers

* lower -> lower -> upper
* capping -> np.where
* Winsorization

```python
df = df[(df['tds']>=lower)&(df['tds']<=upper)]
```

```python
df
```

```python
# capping 
df['tds'] = np.where(df['tds'] > upper,upper,df['tds'])
```

```python
sns.boxplot(data=df, x='tds')
```
